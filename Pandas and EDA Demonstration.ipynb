Basic code to execute histogram to check the data spread for all the store features

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

features  =  ['Store_Area', 'Items_Available', 'Daily_Customer_Count','Store_Sales']

df[features].hist(figsize=(12,10),bins=20,color='blue',edgecolor='black')
plt.suptitle("Histogram for Store Features",fontsize=16)
```

1. Now exectuing a proper histogram to check the skewness and the data spread. If it is normally ditributed.
2. By analysing below graph. we can see skwness is alrignt.

```python

numerical_features = ['Store_Area', 'Items_Available', 'Daily_Customer_Count', 'Store_Sales']

for feature in numerical_features:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[feature], kde=True, bins=20, color='skyblue', edgecolor='black')
    
    # Labels and title
    plt.xlabel(f"{feature}", fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
    plt.title(f"Distribution of {feature}", fontsize=16)
    
    # Show the skewness value
    skewness = df[feature].skew()
    plt.axvline(df[feature].mean(), color='red', linestyle='--', label=f"Mean")
    plt.legend(title=f"Skewness: {skewness:.2f}", loc="upper right")
    
    plt.show()
```

1. Now lets check for Outliers in the data.

```python
# Create box plots for all numerical features
numerical_features = df.select_dtypes(include=['float64', 'int64']).columns

plt.figure(figsize=(12, 7))
for i, feature in enumerate(numerical_features, 1):
    plt.subplot(1, len(numerical_features), i) 
    plt.boxplot(df[feature], vert=True, patch_artist=True, boxprops=dict(facecolor='skyblue'))
    plt.title(f"{feature}", fontsize=10)
    plt.xticks([]) 
    plt.ylabel("Values", fontsize=8)

plt.tight_layout()  
plt.show()
```

1. Doesn't  seems like outliers are greatly affecting the data.
2. Now lets check the correlation.
3. Compute the correlation matrix - "df.corr()"


```python
# Compute the correlation matrix
correlation_matrix = df.corr()


plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix", fontsize=16)
plt.show()
```

1. Creating a pairplot with labels and enhanced visualization  - " sns.pairplot "

```python
# Creating a pairplot with labels and enhanced visualization
sns.pairplot(df, diag_kind="kde", kind="scatter", plot_kws={'alpha': 0.6, 's': 40, 'edgecolor': 'k'})
plt.suptitle("Pairwise Scatter Plot Matrix", y=1.02, fontsize=16)
plt.show()
```

Its hard to interperate anything from correlation graphs as data are circled around mean and "Store_Area" and "Item_Available" have linear relationship between them

```python
# List of features to compare with Store_Sales
features = ['Store_Area', 'Items_Available', 'Daily_Customer_Count']

# Loop through each feature and plot scatter and box plots
for feature in features:
    # Scatter plot
    plt.figure(figsize=(5, 4))
    sns.scatterplot(data=df, x=feature, y='Store_Sales')
    plt.title(f'{feature} vs Store Sales')
    plt.xlabel(feature)
    plt.ylabel('Store Sales')
    plt.show()
```
